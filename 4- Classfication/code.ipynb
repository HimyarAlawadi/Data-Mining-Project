# Naive Bayes Classifier

```python
# Import Wanted Libs
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
# Load dataset and show sample 
df = pd.read_csv('./loan_data.csv')
df.head()
```

```python
# Show Info About The Dataset
df.info()
```

```python
# Show Descriptive statistics 
df.describe()
```

```python
# Visulazition loan data based on purpose of loan
sns.countplot(data=df,x='purpose',hue='not.fully.paid')
plt.xticks(rotation=45, ha='right');
```

```python
# converet purpose column from categorical to integer
pre_df = pd.get_dummies(df,columns=['purpose'],drop_first=True)
pre_df.head()
```

```python
# Define feature that classification depend on it, and split data to training and testing data.
X = pre_df.drop('not.fully.paid', axis=1)
y = pre_df['not.fully.paid']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=125
)
```

```python
# Model building and Training
model = GaussianNB()

model.fit(X_train, y_train);
```

```python
# Model Evluation by accuracy and f1 scores.
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
    classification_report,
)

y_pred = model.predict(X_test)

accuray = accuracy_score(y_pred, y_test)
f1 = f1_score(y_pred, y_test, average="weighted")

print("Accuracy:", accuray)
print("F1 Score:", f1)
```
