# Gaussian Mixture Model

```python
# Import Wanted Lib
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.mixture import GaussianMixture
from sklearn.metrics import accuracy_score
```

```python
# Load the Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
# Visualize the original data
plt.scatter(X[:, 0], X[:, 1], color='blue')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title('Original Data')
plt.show()
```

```python
# Create a GMM object with 3 clusters
gmm = GaussianMixture(n_components=3)
```

```python
# Fit the GMM model to the data
gmm.fit(X)
```

```python
# Predict the cluster labels
labels = gmm.predict(X)
```

```python
# Visualize the clustering
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title('GMM Clustering')
plt.show()
```

```python
# Generate test data
np.random.seed(0)
X_test, y_test = datasets.make_blobs(n_samples=50, centers=3, n_features=4,random_state=0)
```

```python
# Predict the cluster labels for test data
test_labels = gmm.predict(X_test)
```

```python
# Evaluate the clustering results
accuracy = accuracy_score(y_test, test_labels)
```

```python
print("Test Data:")
print("Predicted Labels:", test_labels)
print("True Labels:", y_test)
print("Accuracy:", accuracy)
```
